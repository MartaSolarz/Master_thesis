```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data-old/survey.csv').set_index('ID')
df
```

```python
df.info()
```

```python
df['Płeć'] = df['Płeć'].astype('category')
pd.value_counts(df['Rok_studiów'])
```

```python
def rok_stud(x):
    if x == 'I':
        return 1
    elif x == 'II':
        return 2
    else:
        return 3
    
df['Rok_studiów'] = df['Rok_studiów'].apply(rok_stud)
```

```python
df.info()
```

```python
df
```

```python
pd.unique(df['Specjalność'])
```

```python
df['Specjalność'] = df['Specjalność'].apply(lambda x: np.NaN if x == None else x)
pd.unique(df['Specjalność'])
```

```python
pd.unique(df['Wady_wzroku'])
```

```python
df['Wady_wzroku'] = df['Wady_wzroku'].apply(lambda x: 0 if x == 'nie' else 1)
pd.unique(df['Wady_wzroku'])
```

```python
df['Daltonizm'] = df['Daltonizm'].apply(lambda x: 0 if x == 'nie' else 1)
pd.value_counts(df['Daltonizm'])
```

```python
def sen(x):
    if x == 'mniej niż 6':
        return '<6'
    elif x == 'więcej niż 8':
        return '8<'
    else:
        return x
    
df['Ilość_snu'] = df['Ilość_snu'].apply(sen)
pd.value_counts(df['Ilość_snu'])
```

```python
df
```

```python
df['JP_roz'] = df['JP_roz'].apply(lambda x: np.NaN if x == '-' else x)
df['MAT_roz'] = df['MAT_roz'].apply(lambda x: np.NaN if x == '-' else x)
df['FIZ_roz'] = df['FIZ_roz'].apply(lambda x: np.NaN if x == '-' else x)
df['HIST_roz'] = df['HIST_roz'].apply(lambda x: np.NaN if x == '-' else x)
df
```

```python
def forms(x):
    if x == "A":
        return 'mapa'
    elif x == 'B':
        return 'tabela'
    elif x == 'C':
        return 'wykres'
    else:
        return 'tekst'

df['Quest2'] = df['Quest2'].apply(forms)
pd.value_counts(df['Quest2'])
```

```python
df['Data'] = df['Czas'].apply(lambda x: x[:10])
df
```

```python
df['Godzina'] = df['Czas'].apply(lambda x: x[11:19])
df = df.drop(columns='Czas')
df
```

```python
df.info()
```

```python
df.to_csv('data-new/survey.csv', index=True)
```
