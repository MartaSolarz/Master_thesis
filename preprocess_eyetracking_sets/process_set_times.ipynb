```python
import pandas as pd
```

```python
df = pd.read_csv('data-old/times.tsv', sep='\t')
df
```

```python
df.info()
```

```python
df['ID'] = df['Participant'].apply(lambda x: int(''.join(filter(str.isdigit, x))))
df.loc[df['ID'] == 28, 'ID'] = 29 # only 2B
df
```

```python
df = df.drop(columns=['Recording', 'Participant', 'Interval'])
df
```

```python
def change_name(x):
    if x == 'mapa1_po_poprawce':
        return '1a'
    elif x == 'mapa2_po_poprawce':
        return '2a'
    elif x == 'mapa3_po_poprawce':
        return '3a'
    elif x == 'zadanie_1_po_poprawce':
        return '1b'
    elif x == 'zadanie_2_po_poprawce':
        return '2b'
    elif x == 'zadanie_3_po_poprawce':
        return '3b'
    else:
        return x

df['TOI'] = df['TOI'].apply(change_name)
```

```python
df
```

```python
df.rename(columns={'Duration_of_interval':'Duration', 'Start_of_interval':'Start', 'Last_AOI_viewed': 'Last_AOI'}, inplace=True)
df
```

```python
df_1a = df[df['TOI'] == '1a']
df_1b = df[df['TOI'] == '1b']
df_2a = df[df['TOI'] == '2a']
df_2b = df[df['TOI'] == '2b']
df_3a = df[df['TOI'] == '3a']
df_3b = df[df['TOI'] == '3b']

df_1a
```

```python
df_1a.set_index('ID', inplace=True)
df_1b.set_index('ID', inplace=True)
df_2a.set_index('ID', inplace=True)
df_2b.set_index('ID', inplace=True)
df_3a.set_index('ID', inplace=True)
df_3b.set_index('ID', inplace=True)
```

```python
df_1a = df_1a.add_suffix('_1a')
df_1b = df_1b.add_suffix('_1b')
df_2a = df_2a.add_suffix('_2a')
df_2b = df_2b.add_suffix('_2b')
df_3a = df_3a.add_suffix('_3a')
df_3b = df_3b.add_suffix('_3b')

df_3a
```

```python
df_duration = pd.merge(df_1a, df_1b, on='ID')
df_duration = pd.merge(df_duration, df_2a, on='ID')
df_duration = pd.merge(df_duration, df_2b, on='ID')
df_duration = pd.merge(df_duration, df_3a, on='ID')
df_duration = pd.merge(df_duration, df_3b, on='ID')

df_duration
```

```python
df_duration = df_duration.drop(columns=['TOI_1a', 'TOI_1b', 'TOI_2a', 'TOI_2b', 'TOI_3a', 'TOI_3b'])
df_duration
```
