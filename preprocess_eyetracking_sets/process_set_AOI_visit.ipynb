```python
import pandas as pd
```

```python
NAME = '3b'
df = pd.read_csv('data-old/Poprawki/AOI_visit/3B.tsv', sep='\t')
df
```

```python
df.info()
```

```python
# zmiana nazw kolumn
# usuniÄ™cie recordings
# unikalne indeksy - potem participants i recording, interwal do usuniÄ™cia

df['ID'] = df['Participant'].apply(lambda x: int(''.join(filter(str.isdigit, x))))
df.loc[df['ID'] == 28, 'ID'] = 29 # only 2B
df
```

```python
df = df[df['TOI'] != 'Entire Recording']
df
```

```python
df = df.drop(columns=['Recording', 'Participant', 'TOI', 'Interval'])
df
```

```python
df.rename(columns = {'Total_duration_of_Visit':'Total', 'Average_duration_of_Visit':'Avg', 'Minimum_duration_of_Visit':'Min', 'Maximum_duration_of_Visit':'Max', 'Number_of_Visits':'Num', 'Time_to_first_Visit':'TTV', 'Duration_of_first_Visit':'DTTV'}, inplace=True)
df
```

```python
df_com = df[df['AOI'] == f'{NAME}_com']
df_map = df[df['AOI'] == f'{NAME}_map']
df_plt = df[df['AOI'] == f'{NAME}_plt']
df_tab = df[df['AOI'] == f'{NAME}_tab']
df_txt = df[df['AOI'] == f'{NAME}_txt']
```

```python
df_map.set_index('ID', inplace=True)
df_com.set_index('ID', inplace=True)
df_plt.set_index('ID', inplace=True)
df_tab.set_index('ID', inplace=True)
df_txt.set_index('ID', inplace=True)
```

```python
df_map
```

```python
df_com = df_com.add_suffix(f'_{NAME}_com')
df_map = df_map.add_suffix(f'_{NAME}_map')
df_plt = df_plt.add_suffix(f'_{NAME}_plt')
df_tab = df_tab.add_suffix(f'_{NAME}_tab')
df_txt = df_txt.add_suffix(f'_{NAME}_txt')
df_map
```

```python
df = pd.merge(df_map, df_com, on='ID')
df = pd.merge(df, df_plt, on='ID')
df = pd.merge(df, df_tab, on='ID')
df = pd.merge(df, df_txt, on='ID')
df
```

```python
df = df.drop(columns=[f'AOI_{NAME}_com', f'AOI_{NAME}_map', f'AOI_{NAME}_plt', f'AOI_{NAME}_tab', f'AOI_{NAME}_txt'])
df
```

```python
df.info()
```

```python
df.to_csv(f'data-new/metrics/AOI_visit/{NAME}.csv', index=True)
```
