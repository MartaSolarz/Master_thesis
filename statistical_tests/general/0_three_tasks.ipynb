# 0. Tests for three groups

###  Is there a statistical difference in correctness, response time, and task difficulty assessment between the three tasks? (1 vs 2 vs 3)

Tasks are dependent groups, so the Friedman test will be used to check the differences between the three groups.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from functions.tests.dependent_groups.more_than_two_groups import friedman_test
```

```python
df_answers = pd.read_csv('/Users/martasolarz/Studies/Thesis/Master_thesis/prepare_datasets/sets/correctness.csv')

df_time = pd.read_csv('/Users/martasolarz/Studies/Thesis/Master_thesis/prepare_datasets/sets/times.csv')
```

### 1. Correctness

Sum of correct answers

```python
data = [df_answers['SumCorr_1'], df_answers['SumCorr_2'], df_answers['SumCorr_3']]

friedman_test(data)
```

Correct answers for task without redundancy

```python
data = [df_answers['1a_corr'], df_answers['2a_corr'], df_answers['3a_corr']]

friedman_test(data)
```

Correct answers for task with redundancy

```python
friedman_test([df_answers['1b_corr'], df_answers['2b_corr'], df_answers['3b_corr']])
```

There are no statistically significant differences in correctness for the three groups.

### 2. Response time

Sum of response time

```python
data = [df_time['SumTime_1'], df_time['SumTime_2'], df_time['SumTime_3']]
friedman_test(data)

for d in data:
    sns.kdeplot(d, fill=True)
plt.legend(['1', '2', '3'])
plt.xlabel('Czas odpowiedzi [s]')
plt.grid(True)
plt.show()
```

```python
data = [df_time['Duration_1a'], df_time['Duration_2a'], df_time['Duration_3a']]
friedman_test(data)

for d in data:
    sns.kdeplot(d, fill=True)
plt.legend(['1a', '2a', '3a'])
plt.xlabel('Czas odpowiedzi [s]')
plt.grid(True)
plt.show()
```

```python
data = [df_time['Duration_1b'], df_time['Duration_2b'], df_time['Duration_3b']]
friedman_test(data)

for d in data:
    sns.kdeplot(d, fill=True)
plt.legend(['1b', '2b', '3b'])
plt.xlabel('Czas odpowiedzi [s]')
plt.grid(True)
plt.show()
```

There is a statistically significant relationship in response time for the three graphics for:
- summed values (between graphics 2 and 3 --> 3 longer response time)
- graphic b (between graphics 1 and 3, 2 and 3; --> 3 longer response time in both cases) 

### 3. Task difficulty assessment

Sum of task difficulty assessment

```python
friedman_test([df_answers['SumTrud_1'], df_answers['SumTrud_2'], df_answers['SumTrud_3']])
```

Task difficulty assessment for task without redundancy

```python
friedman_test([df_answers['1a_trud'], df_answers['2a_trud'], df_answers['3a_trud']])
```

Task difficulty assessment for task with redundancy

```python
friedman_test([df_answers['1b_trud'], df_answers['2b_trud'], df_answers['3b_trud']])
```

There is a statistically significant relationship in the assessment of the task difficulty for graphic b. (relationship: 1b with 3b; 2b with 3b)
