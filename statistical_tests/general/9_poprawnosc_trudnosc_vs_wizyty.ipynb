### Testy:

**H0:** Nie ma istotnej statystycznie korelacji między poprawnością odpowiedzi a daną metryką dotyczącą wizyt.

**H1:** Istnieje istotna statystycznie korelacja między poprawnością odpowiedzi a daną metryką dotyczącą wizyt.

**Poziom istotności:** $\alpha=0.05$

```python
ALPHA = 0.05
```

Poprawność vs całkowita, średnia długość wizyty

```python
# sumy grafik
param_category = 'Corr'
for col in cols_continue:
    print('##########################################')
    print(col)
    for i, df in enumerate(dfs_sum):
        print('------------------------------')
        print(names_sum[i+1])
        df = df.dropna()
        correlation_independent_groups_test(df[col], df[param_category], ALPHA)
```

Poprawność vs liczba wizyt

```python
# sumy grafik
param_category = 'Corr'
col = 'Num'
for i, df in enumerate(dfs_sum):
    print('------------------------------')
    print(names_sum[i+1])
    df = df.dropna()
    correlation_independent_groups_test(df[col], df[param_category], ALPHA)
```

Podsumowanie - poprawność:

Istotne statystycznie:

- całkowita długość wizyty: 
    - sumy: 2 (z poleceniem, bez polecenia), 3 (z poleceniem, bez polecenia)
- średnia długość wizyty:
    - sumy: B (z poleceniem), 2 (z poleceniem, bez polecenia), 3 (z poleceniem, bez polecenia)
- liczba wizyt:
    - sumy: A (bez polecenia)
