# Temat testów
– Czy istnieje istotna statystycznie różnica w średniej poprawności odpowiedzi między grupami A vs B?

– Czy istnieje istotna statystycznie różnica w średniej ocenie trudności zadania między grupami A vs B?

– Czy istnieje istotna statystycznie różnica w średnim czasie odpowiedzi między grupami A vs B?

– Czy istnieje istotna statystycznie różnica w średniej całkowitych długości fiksacji/wizyt między grupami A vs B?

– Czy istnieje istotna statystycznie różnica w średniej średnich długości fiksacji/wizyt między grupami A vs B?

– Czy istnieje istotna statystycznie różnica w średniej średnich wielkości źrenicy między grupami A vs B?

– Czy istnieje istotna statystycznie różnica w średniej liczbie fiksacji/wizyt między grupami A vs B?


```python
import pandas as pd

from functions.tests.dependent_groups.two_groups import ttest_or_wilcoxon
from functions.plots import plot_multi_density
```

```python
df_all = pd.read_csv('/Users/martasolarz/Studies/Thesis/Master_thesis/prepare_datasets/sets/analizy_wieloczynnikowe.csv', index_col='idx')
df = df_all.filter(regex='^Avg_(?!fix|visit|pupil)')
df['Group'] = df_all['Group']
df['ID'] = df_all['ID']
```

**H0:** Brak istotnej statystycznie różnicy w średniej wartości X między grupami A i B.

**H1:** Istnieje istotna statystycznie różnica w średniej wartości X między grupami A i B.

**Poziom istotności:** $\alpha$=0.05

```python
ALPHA = 0.05
```

```python
df.head()
```

```python
# poprawność
ttest_or_wilcoxon([df['Avg_Corr'][df['Group'] == 'A'], df['Avg_Corr'][df['Group'] == 'B']], ALPHA)
```

```python
# trudność
ttest_or_wilcoxon([df['Avg_Diff'][df['Group'] == 'A'], df['Avg_Diff'][df['Group'] == 'B']], ALPHA)
```

```python
# czas
ttest_or_wilcoxon([df['Avg_Time'][df['Group'] == 'A'], df['Avg_Time'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_Time', 'Group', 'Czas [ms]', 'Grupa grafik')
```

```python
# całkowita długości fiksacji
ttest_or_wilcoxon([df['Avg_TotalFix_WithCom'][df['Group'] == 'A'], df['Avg_TotalFix_WithCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_TotalFix_WithCom', 'Group', 'Całkowita długość fiksacji (z poleceniem) [ms]', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_TotalFix_WithoutCom'][df['Group'] == 'A'], df['Avg_TotalFix_WithoutCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_TotalFix_WithoutCom', 'Group', 'Całkowita długość fiksacji (bez polecenia) [ms]', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_TotalFix_OnlyCom'][df['Group'] == 'A'], df['Avg_TotalFix_OnlyCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_TotalFix_OnlyCom', 'Group', 'Całkowita długość fiksacji (tylko polecenie) [ms]', 'Grupa grafik')
```

```python
# średnia długość fiksacji
ttest_or_wilcoxon([df['Avg_AvgFix_WithCom'][df['Group'] == 'A'], df['Avg_AvgFix_WithCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_AvgFix_WithCom', 'Group', 'Średnia długość fiksacji (z poleceniem) [ms]', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_AvgFix_WithoutCom'][df['Group'] == 'A'], df['Avg_AvgFix_WithoutCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_AvgFix_WithoutCom', 'Group', 'Średnia długość fiksacji (bez polecenia) [ms]', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_AvgFix_OnlyCom'][df['Group'] == 'A'], df['Avg_AvgFix_OnlyCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_AvgFix_OnlyCom', 'Group', 'Średnia długość fiksacji (tylko polecenie) [ms]', 'Grupa grafik')
```

```python
# liczba fiksacji
ttest_or_wilcoxon([df['Avg_NumFix_WithCom'][df['Group'] == 'A'], df['Avg_NumFix_WithCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_NumFix_WithCom', 'Group', 'Liczba fiksacji (z poleceniem)', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_NumFix_WithoutCom'][df['Group'] == 'A'], df['Avg_NumFix_WithoutCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_NumFix_WithoutCom', 'Group', 'Liczba fiksacji (bez polecenia)', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_NumFix_OnlyCom'][df['Group'] == 'A'], df['Avg_NumFix_OnlyCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_NumFix_OnlyCom', 'Group', 'Liczba fiksacji (tylko polecenie)', 'Grupa grafik')
```

```python
# średnia wielkość źrenicy
ttest_or_wilcoxon([df['Avg_AvgPupil_WithCom'][df['Group'] == 'A'], df['Avg_AvgPupil_WithCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_AvgPupil_WithCom', 'Group', 'Średnia wielkość źrenicy (z poleceniem)', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_AvgPupil_WithoutCom'][df['Group'] == 'A'], df['Avg_AvgPupil_WithoutCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_AvgPupil_WithoutCom', 'Group', 'Średnia wielkość źrenicy (bez polecenia)', 'Grupa grafik')
print('----------')
a = df[['Avg_AvgPupil_OnlyCom', 'ID']][df['Group'] == 'A']
b = df[['Avg_AvgPupil_OnlyCom', 'ID']][df['Group'] == 'B']
ab = a.merge(b, on='ID')
ab = ab.dropna()
ttest_or_wilcoxon([ab['Avg_AvgPupil_OnlyCom_x'], ab['Avg_AvgPupil_OnlyCom_y']], ALPHA)
plot_multi_density(df, 'Avg_AvgPupil_OnlyCom', 'Group', 'Średnia wielkość źrenicy (tylko polecenie)', 'Grupa grafik')
```

```python
# całkowita długość wizyt
ttest_or_wilcoxon([df['Avg_TotalVisit_WithCom'][df['Group'] == 'A'], df['Avg_TotalVisit_WithCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_TotalVisit_WithCom', 'Group', 'Całkowita długość wizyt (z poleceniem) [ms]', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_TotalVisit_WithoutCom'][df['Group'] == 'A'], df['Avg_TotalVisit_WithoutCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_TotalVisit_WithoutCom', 'Group', 'Całkowita długość wizyt (bez polecenia) [ms]', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_TotalVisit_OnlyCom'][df['Group'] == 'A'], df['Avg_TotalVisit_OnlyCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_TotalVisit_OnlyCom', 'Group', 'Całkowita długość wizyt (tylko polecenie) [ms]', 'Grupa grafik')
```

```python
# średnia długość wizyt
ttest_or_wilcoxon([df['Avg_AvgVisit_WithCom'][df['Group'] == 'A'], df['Avg_AvgVisit_WithCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_AvgVisit_WithCom', 'Group', 'Średnia długość wizyt (z poleceniem) [ms]', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_AvgVisit_WithoutCom'][df['Group'] == 'A'], df['Avg_AvgVisit_WithoutCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_AvgVisit_WithoutCom', 'Group', 'Średnia długość wizyt (bez polecenia) [ms]', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_AvgVisit_OnlyCom'][df['Group'] == 'A'], df['Avg_AvgVisit_OnlyCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_AvgVisit_OnlyCom', 'Group', 'Średnia długość wizyt (tylko polecenie) [ms]', 'Grupa grafik')
```

```python
# liczba wizyt
ttest_or_wilcoxon([df['Avg_NumVisit_WithCom'][df['Group'] == 'A'], df['Avg_NumVisit_WithCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_NumVisit_WithCom', 'Group', 'Liczba wizyt (z poleceniem)', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_NumVisit_WithoutCom'][df['Group'] == 'A'], df['Avg_NumVisit_WithoutCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_NumVisit_WithoutCom', 'Group', 'Liczba wizyt (bez polecenia)', 'Grupa grafik')
print('----------')
ttest_or_wilcoxon([df['Avg_NumVisit_OnlyCom'][df['Group'] == 'A'], df['Avg_NumVisit_OnlyCom'][df['Group'] == 'B']], ALPHA)
plot_multi_density(df, 'Avg_NumVisit_OnlyCom', 'Group', 'Liczba wizyt (tylko polecenie)', 'Grupa grafik')
```
