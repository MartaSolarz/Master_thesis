```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./sets/correctness.csv')
df
```

```python
def create_barplot(df, col, label, maxvalue=21, flag=False):
    data = df[col].value_counts().sort_index()
    if flag:
        data = data.append(pd.Series([0], index=[5]))

    plt.figure(figsize=(5, 4))
    bars = plt.bar(data.index, data, color='skyblue', edgecolor='black', width=0.5)

    plt.title(f'Ocena trudności zadania dla {label}\n($min=0$; $max=5$)')
    plt.xlabel(f'Ocena trudności zadania dla {label}')
    plt.ylabel('Liczba osób')
    plt.xticks(data.index)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.ylim(0, maxvalue)

    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                 str(int(bar.get_height())), ha='center', va='bottom')
        
    plt.yticks(range(0, maxvalue, 5))

    plt.show()
```

```python
create_barplot(df, '1a_trud', 'grafiki 1A')
create_barplot(df, '1b_trud', 'grafiki 1B', 21, True)

create_barplot(df, '2a_trud', 'grafiki 2A')
create_barplot(df, '2b_trud', 'grafiki 2B')

create_barplot(df, '3a_trud', 'grafiki 3A')
create_barplot(df, '3b_trud', 'grafiki 3B')
```

```python
plt.figure(figsize=(10, 5))

mean_1 = df['1a_trud'].mean()
mean_3 = df['2a_trud'].mean()
mean_5 = df['3a_trud'].mean()

plt.plot(df['1a_trud'], label=f'Grafika 1A ($\mu={mean_1:.2f}$)', linestyle='-', color='blue')

plt.plot(df['2a_trud'], label=f'Grafika 2A ($\mu={mean_3:.2f}$)', linestyle='-', color='green')

plt.plot(df['3a_trud'], label=f'Grafika 3A ($\mu={mean_5:.2f}$)', linestyle='-', color='red')
plt.axhline(y=mean_1, color='blue', linestyle=':')
plt.axhline(y=mean_3, color='green', linestyle=':')
plt.axhline(y=mean_5, color='red', linestyle=':')

plt.title('Ocena trudności zadania przez poszczególnych uczestników - grafiki typu A ($min=0$; $max=5$)')
plt.xlabel('Uczestnik badania')
plt.ylabel('Ocena trudności zadania')

plt.yticks(range(0, 6, 1))
plt.legend()
plt.grid(True)
plt.show()
```

```python
def create_barplot_corr(df, col, label, maxvalue=40, flag=False):
    data = df[col].value_counts().sort_index()
    if flag:
        data = data.append(pd.Series([0], index=[5]))

    plt.figure(figsize=(4, 4))
    bars = plt.bar(data.index, data, color='skyblue', edgecolor='black', width=0.5)

    plt.title(f'Poprawność odpowiedzi dla {label}\n($poprawna=1$; $błędna=0$)', size=11)
    plt.xlabel(f'Poprawność odpowiedzi')
    plt.ylabel('Liczba osób')
    plt.xticks(data.index)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.ylim(0, maxvalue)

    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                 str(int(bar.get_height())), ha='center', va='bottom')
        
    plt.yticks(range(0, maxvalue, 5))

    plt.show()
```

```python
create_barplot_corr(df, '1a_corr', 'grafiki 1A')
create_barplot_corr(df, '1b_corr', 'grafiki 1B')

create_barplot_corr(df, '2a_corr', 'grafiki 2A')
create_barplot_corr(df, '2b_corr', 'grafiki 2B')

create_barplot_corr(df, '3a_corr', 'grafiki 3A')
create_barplot_corr(df, '3b_corr', 'grafiki 3B')
```
